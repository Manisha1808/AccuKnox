```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Manisha1808/AccuKnox/main/StudentsPerformance.csv"
df = pd.read_csv(url)
```

```python
df["average_score"] = (df["math score"] + df["reading score"] + df["writing score"]) / 3
print("Average of all students:", df["average_score"].mean())
```

```python
plt.bar(df.index, df["average_score"])
plt.xlabel("Students")
plt.ylabel("Average Score")
plt.title("Avg Test Scores")
plt.show()
```
